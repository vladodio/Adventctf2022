```python
lines = []

with open("in3.txt", "r") as file:
    lines = file.readlines()
lines = [i.strip() for i in lines]

lines[:10]
```

```python
import string
val_lookup = {}
counter = 1
for i in string.ascii_lowercase:
    val_lookup[i] = counter
    counter += 1
    
for i in string.ascii_uppercase:
    val_lookup[i] = counter
    counter += 1
val_lookup
```

```python
out_lst = []
for line in lines:
    a_set = set(line[:len(line)//2])
    b_set = set(line[len(line)//2:])
    out_lst.append(  list(a_set.intersection(b_set)) )
out_lst = [item for sublist in out_lst for item in sublist]
out_lst
```

```python
msum = 0
for item in out_lst:
    msum += val_lookup[item]
msum
```

```python
out_lst = []
for i in range(0, len(lines) - 2, 3):
    out_lst = out_lst + list(set(lines[i]).intersection(set(lines[i+1]).intersection(set(lines[i+2]))))
out_lst
```

```python
msum = 0
for item in out_lst:
    msum += val_lookup[item]
msum
```

## Revised Solution

```python
out_lst = []
for line in lines:
    a_set = set(line[:len(line)//2])
    b_set = set(line[len(line)//2:])
    out_lst += list(a_set.intersection(b_set))

sum([val_lookup[item] for item in out_lst])
```

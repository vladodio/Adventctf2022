```python
lines = []

with open("in5.txt", "r") as file:
    lines = file.readlines()
lines = [i.rstrip() for i in lines]

lines[:10]
```

```python
instructions = lines[lines.index('')+1:]
instructions = [ i.replace("move ", "").replace(" from ", " ").replace(" to ", " ") for i in instructions]
instructions = [ [int(j) for j in i.split()] for i in instructions]
instructions[:10]
```

```python
struct = lines[:lines.index('')]
width = int(struct[-1].split(" ")[-1])
struct = [ i.replace("    ", " ?").replace("[", "").replace("]", "").split(" ") for i in struct]

struct = struct[:-1]
struct
```

```python
columns = [[] for i in range(width) ]

for i in range(len(struct)):
    for j in range(len(struct[i])):
        if(struct[i][j] != "?"):
            columns[j].insert(0, struct[i][j])
columns
```

```python
def process_instruction(instruction):
    amount,start,end = instruction
    print(instruction)
    temp = columns[start-1][len(columns[start-1])-amount:] 
    columns[start-1] = columns[start-1][:len(columns[start-1])-amount]
    #temp.reverse()
    columns[end-1] = columns[end-1] + temp
    print(temp)
    
for i in instructions:
    process_instruction(i)
```

```python
output = ''.join([i[-1] for i in columns])
output
```
